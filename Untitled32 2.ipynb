```python
import pandas as pd

train = pd.read_csv("mitbih_train.csv", header=None)
test = pd.read_csv("mitbih_test.csv", header=None)

print(train.shape)   # should be like (87554, 188)
```

```python
train['label'] = train[187].apply(lambda x: 0 if x == 0 else 1)
test['label'] = test[187].apply(lambda x: 0 if x == 0 else 1)
```

```python
import os

os.makedirs("dataset/train/normal", exist_ok=True)
os.makedirs("dataset/train/abnormal", exist_ok=True)
os.makedirs("dataset/test/normal", exist_ok=True)
os.makedirs("dataset/test/abnormal", exist_ok=True)
```

```python
import matplotlib.pyplot as plt

for i in range(len(train)):
    signal = train.iloc[i, :-2].values   # ECG values
    label = train.iloc[i, -1]

    plt.figure(figsize=(4,2))
    plt.plot(signal)
    plt.axis('off')

    if label == 0:
        plt.savefig(f"dataset/train/normal/{i}.png", bbox_inches='tight')
    else:
        plt.savefig(f"dataset/train/abnormal/{i}.png", bbox_inches='tight')

    plt.close()
```

```python
for i in range(len(test)):
    signal = test.iloc[i, :-2].values
    label = test.iloc[i, -1]

    plt.figure(figsize=(4,2))
    plt.plot(signal)
    plt.axis('off')

    if label == 0:
        plt.savefig(f"dataset/test/normal/{i}.png", bbox_inches='tight')
    else:
        plt.savefig(f"dataset/test/abnormal/{i}.png", bbox_inches='tight')

    plt.close()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_data = datagen.flow_from_directory(
    'dataset/train',
    target_size=(224,224),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

val_data = datagen.flow_from_directory(
    'dataset/train',
    target_size=(224,224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10
)
```

```python
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=5
)
```

```python
test_gen = ImageDataGenerator(rescale=1./255)

test_data = test_gen.flow_from_directory(
    'dataset/test',
    target_size=(224,224),
    batch_size=32,
    class_mode='binary'
)

model.evaluate(test_data)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

img = image.load_img("/content/dataset/test/normal/1000.png", target_size=(224,224))
img_array = np.expand_dims(image.img_to_array(img)/255, axis=0)

pred = model.predict(img_array)

print("Abnormal" if pred > 0.5 else "Normal")
```
